```python
import pandas as pd
```

## Import Data and preproccess it before dealing with P Value

```python
basesurveydf = pd.read_csv('data/BaseSurveyData.csv')

#Get the average complete time
basesurveyAverageCompleteTime = basesurveydf['Duration (in seconds)'].drop([0,1])
basesurveyAverageCompleteTime= basesurveyAverageCompleteTime.astype(int).mean()
print(basesurveyAverageCompleteTime)

#Drop unnecessary columns
basesurveydf.drop(['StartDate', 'EndDate', 'Status', 'IPAddress','Progress', 'Finished', 'RecordedDate', 'RecipientLastName', 'RecipientFirstName', 'RecipientEmail', 'ExternalReference' \
                   ,'LocationLatitude',	'LocationLongitude', 'DistributionChannel', 'UserLanguage', 'Duration (in seconds)', 'ResponseId', 'RandomSurveyID'], axis=1, inplace=True)
basesurveydf.drop([1],inplace=True)

#Make Q3 something we can actually read, probably shouldnt have done this question tbf
q3_options = ['1', '2', '3', '4', '5']
for option in q3_options:
    basesurveydf[f'Q3_option_{option}'] = basesurveydf['Q3'].apply(lambda x: 1 if option in str(x).split(',') else 0)

basesurveydf.drop(columns=['Q3'],inplace=True)

#Remove question
basesurveydf.drop([0], inplace=True)

#Display new df
basesurveydf
```

```python
rearrangedSuveydf = pd.read_csv('data/RearrangeSurveyData.csv')

#Get the average complete time
rearrangedSuveyAverageCompleteTime = rearrangedSuveydf['Duration (in seconds)'].drop([0,1])
rearrangedSuveyAverageCompleteTime= rearrangedSuveyAverageCompleteTime.astype(int).mean()
print(rearrangedSuveyAverageCompleteTime)

#Drop unnecessary columns
rearrangedSuveydf.drop(['StartDate', 'EndDate', 'Status', 'IPAddress','Progress', 'Finished', 'RecordedDate', 'RecipientLastName', 'RecipientFirstName', 'RecipientEmail', 'ExternalReference' \
                   ,'LocationLatitude',	'LocationLongitude', 'DistributionChannel', 'UserLanguage', 'Duration (in seconds)', 'ResponseId', 'RandomSurveyID'], axis=1, inplace=True)
rearrangedSuveydf.drop([1],inplace=True)

#Rearrange and rename questions to be in order like the others
rearrangedSuveydf[['Q1', 'Q2', 'Q3', 'Q4', 'Q5_1', 'Q6_1', 'Q7_1', 'Q8_1', 'Q8_2', 'Q8_3', 'Q8_4', 'Q9_1', 'Q9_2', 'Q9_3', 'Q9_4', 'Q10']] = rearrangedSuveydf[['Q4', 'Q2', 'Q1', 'Q3', 'Q7_1', 'Q6_1', 'Q5_1', 'Q10', 'Q9_1', 'Q9_2', 'Q9_3', 'Q9_4', 'Q8_1', 'Q8_2', 'Q8_3', 'Q8_4']]

rearrangedSuveydf = rearrangedSuveydf.rename({'Q8_1': 'Q8', 'Q8_2': 'Q9_1', 'Q8_3': 'Q9_2', 'Q8_4': 'Q9_3', 'Q9_1': 'Q9_4', 'Q9_2': 'Q10_1', 'Q9_3': 'Q10_2', 'Q9_4': 'Q10_3', 'Q10': 'Q10_4'}, axis='columns')

'Q1', 'Q2', 'Q3', 'Q4', 'Q5_1', 'Q6_1', 'Q7_1', 'Q8', 'Q9_1', 'Q9_2', 'Q9_3', 'Q9_4', 'Q10_1', 'Q10_2', 'Q10_3', 'Q10_4'
#Make Q3 something we can actually read, probably shouldnt have done this question tbf
q3_options = ['1', '2', '3', '4', '5']
for option in q3_options:
    rearrangedSuveydf[f'Q3_option_{option}'] = rearrangedSuveydf['Q3'].apply(lambda x: 1 if option in str(x).split(',') else 0)

rearrangedSuveydf.drop(columns=['Q3'],inplace=True)

#Remove question
rearrangedSuveydf.drop([0], inplace=True)

#Display df
rearrangedSuveydf
```

```python
extendedSurveydf = pd.read_csv('data/ExtendedSurveyData.csv')

#Get average completion time
extendedSuveyAverageCompleteTime = extendedSurveydf['Duration (in seconds)'].drop([0,1])
extendedSuveyAverageCompleteTime= extendedSuveyAverageCompleteTime.astype(int).mean()
print(extendedSuveyAverageCompleteTime)

#Drop unnecessary columns
extendedSurveydf.drop(['StartDate', 'EndDate', 'Status', 'IPAddress','Progress', 'Finished', 'RecordedDate', 'RecipientLastName', 'RecipientFirstName', 'RecipientEmail', 'ExternalReference' \
                   ,'LocationLatitude',	'LocationLongitude', 'DistributionChannel', 'UserLanguage', 'Duration (in seconds)', 'ResponseId', 'RandomSurveyID'], axis=1, inplace=True)
extendedSurveydf.drop([1], inplace=True)

#Make Q3 something we can actually read, probably shouldnt have done this question tbf
q3_options = ['1', '2', '3', '4', '5']
for option in q3_options:
    extendedSurveydf[f'Q3_option_{option}'] = extendedSurveydf['Q3'].apply(lambda x: 1 if option in str(x).split(',') else 0)

extendedSurveydf.drop(columns=['Q3'],inplace=True)

#Remove question
extendedSurveydf.drop([0], inplace=True)

#Display df
extendedSurveydf
```

## P Values

```python
from scipy.stats import ttest_ind

# Flatten responses from all questions for each survey
survey1_responses = basesurveydf.values.flatten().astype(int)
survey2_responses = extendedSurveydf.values.flatten().astype(int)
survey3_responses = rearrangedSuveydf.values.flatten().astype(int)

# Perform T-Test for Survey 1 vs Survey 2
t_stat1, p_value1 = ttest_ind(survey1_responses, survey2_responses)

# Perform T-Test for Survey 1 vs Survey 3
t_stat2, p_value2 = ttest_ind(survey1_responses, survey3_responses)

# Perform T-Test for Survey 2 vs Survey 3
t_stat3, p_value3 = ttest_ind(survey2_responses, survey3_responses)

# Print the results
print(f"T-Test P-value for Survey 1 vs Survey 2: {p_value1}")
print(f"T-Test P-value for Survey 1 vs Survey 3: {p_value2}")
print(f"T-Test P-value for Survey 2 vs Survey 3: {p_value3}")
```
