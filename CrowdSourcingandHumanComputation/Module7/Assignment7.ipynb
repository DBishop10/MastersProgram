```python
import pandas as pd
import pingouin as pg
```

```python
conditionA = pd.read_csv('data/Condition_A.csv')

conditionA.drop(['RecipientFirstName', 'RecipientLastName', 'RecipientEmail','ExternalReference'], axis=1, inplace=True)

conditionA.dropna(inplace=True)

#Get the average complete time
ConditionAAverageCompleteTime = conditionA['Duration (in seconds)'].drop([0,1])
ConditionAAverageCompleteTime= ConditionAAverageCompleteTime.astype(int).mean()
print(ConditionAAverageCompleteTime)

# Drop unnecessary columns
conditionA.drop(['StartDate', 'EndDate', 'Status', 'IPAddress','Progress', 'Finished', 'RecordedDate', \
                   'LocationLatitude',	'LocationLongitude', 'DistributionChannel', 'UserLanguage', 'Duration (in seconds)', 'ResponseId', 'RandomSurveyID'], axis=1, inplace=True)
conditionA.drop([1],inplace=True)


#Remove question
conditionA.drop([0], inplace=True)

#Display new df
conditionA
```

```python
conditionB = pd.read_csv('data/Condition_B.csv')

conditionB.drop(['RecipientFirstName', 'RecipientLastName', 'RecipientEmail','ExternalReference'], axis=1, inplace=True)

conditionB.dropna(inplace=True)


#Get the average complete time
conditionBAverageCompleteTime = conditionB['Duration (in seconds)'].drop([0,1])
conditionBAverageCompleteTime= conditionBAverageCompleteTime.astype(int).mean()
print(conditionBAverageCompleteTime)

#Drop unnecessary columns
conditionB.drop(['StartDate', 'EndDate', 'Status', 'IPAddress','Progress', 'Finished', 'RecordedDate' \
                   ,'LocationLatitude',	'LocationLongitude', 'DistributionChannel', 'UserLanguage', 'Duration (in seconds)', 'ResponseId', 'RandomSurveyID'], axis=1, inplace=True)
conditionB.drop([1],inplace=True)

#Remove question
conditionB.drop([0], inplace=True)



#Display new df
conditionB
```

```python
# Bogardus Social distance scale

bogardus_columns = ['Q1#1_1', 'Q1#1_2',	'Q1#1_3', 'Q1#1_4',	'Q1#1_5', 'Q1#1_6',	'Q1#1_7']

conditionABogardus = conditionA[bogardus_columns].apply(pd.to_numeric, errors='coerce')
conditionBBogardus = conditionB[bogardus_columns].apply(pd.to_numeric, errors='coerce')

# Flatten responses from all questions for each survey
conditionABogardus = conditionABogardus.values.flatten().astype(int)
conditionBBogardus = conditionBBogardus.values.flatten().astype(int)

# Perform T-Test for Survey 1 vs Survey 2
t_test = pg.ttest(conditionABogardus, conditionBBogardus, correction=False)

# Print the results
print(f"T-statistic: {t_test}")
```

```python

# Assuming the perceived dangerousness questions are from Q2_1 to Q2_6
dangerousness_columns = ['Q2_1', 'Q2_2', 'Q2_3', 'Q2_4', 'Q2_5', 'Q2_6']

# Extract the columns for Condition A and B
conditionADangerousness = conditionA[dangerousness_columns].apply(pd.to_numeric, errors='coerce')
conditionBDangerousness = conditionB[dangerousness_columns].apply(pd.to_numeric, errors='coerce')

alpha_A = pg.cronbach_alpha(conditionADangerousness)
alpha_B = pg.cronbach_alpha(conditionBDangerousness)

print(f"Cronbach's Alpha for Condition A: {alpha_A}")
print(f"Cronbach's Alpha for Condition B: {alpha_B}")

# 2. Calculate average response per worker
conditionA_avg = conditionADangerousness.mean(axis=1)
conditionB_avg = conditionBDangerousness.mean(axis=1)

# 3. Perform T-Test for perceived dangerousness scores
t_test = pg.ttest(conditionA_avg, conditionB_avg, correction=False)

# Print the T-test results
print(f"T-statistic: {t_test}")
```

```python
# Social distance scale questions based on the Jim Johnson vignette
social_distance_columns = ['Q3_1', 'Q3_2', 'Q3_3', 'Q3_4', 'Q3_5', 'Q3_6', 'Q3_7', 'Q3_8']

# Extract the columns for Condition A and Condition B
conditionASocialDistance = conditionA[social_distance_columns].apply(pd.to_numeric, errors='coerce')
conditionBSocialDistance = conditionB[social_distance_columns].apply(pd.to_numeric, errors='coerce')

# 1. Calculate Cronbach's Alpha for internal consistency using pingouin
alpha_A_social = pg.cronbach_alpha(conditionASocialDistance)
alpha_B_social = pg.cronbach_alpha(conditionBSocialDistance)

print(f"Cronbach's Alpha for Condition A (Social Distance): {alpha_A_social}")
print(f"Cronbach's Alpha for Condition B (Social Distance): {alpha_B_social}")

# 2. Calculate average response per worker
conditionA_social_avg = conditionASocialDistance.mean(axis=1)
conditionB_social_avg = conditionBSocialDistance.mean(axis=1)

# 3. Perform T-Test using pingouin
t_test = pg.ttest(conditionA_social_avg, conditionB_social_avg, correction=False)

# Print the T-test results
print(t_test)
```

```python
conditionA_bogardus_avg = conditionABogardus.reshape(-1, len(bogardus_columns)).mean(axis=1)
conditionB_bogardus_avg = conditionBBogardus.reshape(-1, len(bogardus_columns)).mean(axis=1)

conditionA_df = pd.DataFrame({
    'Bogardus_Social_Distance': conditionA_bogardus_avg,
    'Perceived_Dangerousness': conditionA_avg,
    'Jim_Johnson_Social_Distance': conditionA_social_avg
})

conditionB_df = pd.DataFrame({
    'Bogardus_Social_Distance': conditionB_bogardus_avg,
    'Perceived_Dangerousness': conditionB_avg,
    'Jim_Johnson_Social_Distance': conditionB_social_avg
})

# Step 2: Calculate correlation matrices for each condition
correlation_matrix_A = conditionA_df.corr()
correlation_matrix_B = conditionB_df.corr()

# Print the correlation matrices
print("Correlation Matrix for Condition A:")
print(correlation_matrix_A)

print("\nCorrelation Matrix for Condition B:")
print(correlation_matrix_B)
```
