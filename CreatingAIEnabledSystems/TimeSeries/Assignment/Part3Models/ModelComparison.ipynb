## Final Thoughts on these models compared to part 1

Overall both of the parts had models that performed incredibly well with the dataset provided. I believe the reason for this is because the total transactions is fairly regular and follows an easily noticable trend over the year. To fully look at the differences between part 1 and 3 models I have taken the exact predicted values for both models and compared them to the actual value to see who is overall closest.

```python
import os
os.chdir('../../')

import pandas as pd
import numpy as np

p1dm = pd.read_csv('p1dm.csv')
p1tm = pd.read_csv('p1tm.csv')
p3dm = pd.read_csv('p3dm.csv')
p3tm = pd.read_csv('p3tm.csv')
```

```python
p3tm.head()
```

```python
P1Deepdif = abs(p1dm['Actual_Transactions'] - p1dm['Predicted_Transactions']).sum()
P1Traddif = abs(p1tm['Actual_Transactions'] - p1tm['Predicted_Transactions']).sum()
P3Deepdif = abs(p3dm['Actual_Transactions'] - p3dm['Predicted_Transactions']).sum()
P3Traddif = abs(p3tm['Actual_Transactions'] - p3tm['Predicted_Transactions']).sum()

print(f"The total absolute difference for Deep Model 1 is: {P1Deepdif}")
print(f"The total absolute difference for Deep Model 3 is: {P3Deepdif}")
print(f"The total absolute difference for Traditional Model 1 is: {P1Traddif}")
print(f"The total absolute difference for Traditional Model 3 is: {P3Traddif}")
```

As we can see these models overall are all incredibly accurate. Somethinginteresting to point out is that, while the Traditional Model for part 3 performed insanely well compared to part 1, the deep model for part 3 actual performed slightly worse overall compared to the deep model for part 1
