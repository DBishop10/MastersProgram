## Findings When determining transactions and fraudulent transactions based on a sub-feature

The subfeature I ended up going with was daily_transaction_amount. I felt that this would correlate well into amount of daily fraud as well as just transactions overall. I ran a few tests with both the traditional and deep model and found that, much like the graphs in general they were both fairly good at determining daily transactions but struggled to get very close to fraudulent transactions. I also ran into the issue that the model couldn't be given just the daily_transaction_amount to forecast. I am unsure if this is what I was supposed to do but the easiest way I found to get around this was to utilize an averaged historical data. This way I found a row that had a similar daily_transaction_amount, replaced it with the transaction amount that was inputted and then fed into the models. I personally believe this is still using the sub feature to determine the number of transactions and fraudulent transactions.

As we can see these models overall are all incredibly accurate. Somethinginteresting to point out is that, while the Traditional Model for part 3 performed insanely well compared to part 1, the deep model for part 3 actual performed slightly worse overall compared to the deep model for part 1
