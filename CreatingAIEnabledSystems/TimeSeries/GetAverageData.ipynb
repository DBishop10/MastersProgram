```python
import pandas as pd
import numpy as np
from data_pipeline import Pipeline
```

```python
data = pd.read_csv('data/CreditCardFraudFourYears.csv')
```

```python
pipeline = Pipeline()
transformed_df = pipeline.transform(data)
```

```python
for column in transformed_df.columns:
    if pd.api.types.is_bool_dtype(transformed_df[column]):
        transformed_df[column] = transformed_df[column].astype(int)
    if pd.api.types.is_integer_dtype(transformed_df[column]):
        transformed_df[column] = transformed_df[column].astype(float)
transformed_df.dtypes
```

```python
transformed_df['mm_dd'] = transformed_df.index.strftime('%m/%d')

transformed_df.head()
```

```python
average_daily_data = transformed_df.groupby('mm_dd').mean().reset_index()
average_daily_data.set_index('mm_dd', inplace=True)
```

```python
average_daily_data.to_csv('average_daily_data.csv')
```
